### Executive Summary

> The strongest predictors of future giving are prior giving behavior and contact activity. While event attendance contributes positively, donor type, relationship volume, and the synthetic engagement score provide little additional explanatory power. These findings suggest that consistent stewardship and prior philanthropic behavior are more important indicators of future giving than constituent category or network size.

````python
#k means cluster
cluster_cols = [
    'Lifetime_Giving',
    'giving_total_gifts',
    'contact_count',
    'events_attended',
    'Will_Give_Again_Probability'
]

from sklearn.preprocessing import StandardScaler

X = model_df[cluster_cols].fillna(0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(2, 11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(2,11), inertia, marker='o')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
````

````python
kmeans = KMeans(
    n_clusters=5,
    random_state=42,
    n_init=10
)

model_df['cluster'] = kmeans.fit_predict(X_scaled)

cluster_profile = model_df.groupby('cluster')[cluster_cols].mean().round(2)

print(cluster_profile)

print(model_df['cluster'].value_counts())
````

````python
# Relationships

relationship_features = relationships.groupby('Donor_ID_1').agg(
    relationship_count=('Donor_ID_2', 'count'),
    avg_relationship_strength=('Relationship_Strength', 'mean'),
    max_relationship_strength=('Relationship_Strength', 'max'),
    relationship_type_count=('Relationship_Type', 'nunique')
).reset_index()
````

````python
# Contacts

contact_features = contacts.groupby('Donor_ID').agg(
    contact_count=('Contact_Report_ID', 'count'),
    contact_type_count=('Contact_Type', 'nunique')
).reset_index()
````

````python
model_df = donor_master.copy()

model_df = model_df.merge(giving_features, left_on='ID', right_on='Donor_ID', how='left')
model_df = model_df.drop(columns=['Donor_ID'])

model_df = model_df.merge(contact_features, left_on='ID', right_on='Donor_ID', how='left')
model_df = model_df.drop(columns=['Donor_ID'])

model_df = model_df.merge(event_features, left_on='ID', right_on='Donor_ID', how='left')
model_df = model_df.drop(columns=['Donor_ID'])

model_df = model_df.merge(relationship_features, left_on='ID', right_on='Donor_ID_1', how='left')
model_df = model_df.drop(columns=['Donor_ID_1'])

model_df = model_df.rename(columns={
    'total_gifts_y': 'giving_total_gifts',
    'avg_relationship_strength_y': 'rel_avg_strength',
    'max_relationship_strength_y': 'rel_max_strength'
})

feature_cols = [
    'giving_total_gifts',
    'avg_gift',
    'max_gift',
    'total_given',
    'contact_count',
    'contact_type_count',
    'events_attended',
    'unique_events',
    'relationship_count',
    'rel_avg_strength',
    'rel_max_strength',
    'relationship_type_count'
]

model_df[feature_cols] = model_df[feature_cols].fillna(0)

print(model_df.shape)
````
